```python
import numpy as np
import matplotlib.pyplot as plt
import gif

M = 10000
```

**Función de Weiestrass**

$f(x) = \sum_{n=0}^{\infty}a^n\cos(b^n \pi x) $, 

donde 

$0 < a < 1$, $b$ es un entero positivo impar y $ab > 1 + \frac{3}{2}\pi$.

```python
def w(x, a=0.5, b=23, n=100):
    "Función de Weierstrass"
    
    exp = np.arange(0, n)
    a_n = np.power(a, exp)
    b_n = np.power(b, exp)
    return np.dot(a_n, np.cos(b_n * np.pi * x))
```

```python
vw = np.vectorize(w)
```

```python
x = np.linspace(-2,2,M)
y = vw(x, b = 15)
fig= plt.figure(figsize=(12,6))
plt.plot(x, y, 'k-', lw=0.2)
plt.title('Función de Weierstrass para b = 15 y a = 0.5')
plt.show()
```

Creación de gif:

```python
@gif.frame
def plot(b):
    x = np.linspace(-2,2,M)
    y_ = np.repeat(-2, x.shape[0])
    y = vw(x, b=b)
    plt.figure(figsize=(12,6))
    plt.plot(x, y, 'k-', lw=0.3)
    plt.axis('off')
```

```python
frames = []
for b in np.linspace(1, 5, 60):
    frame = plot(b)
    frames.append(frame)
```

```python
gif.save(frames, 'weierstrass.gif', duration=10)
```
